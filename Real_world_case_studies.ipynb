```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Generate synthetic customer data (Income and Spending Score)
np.random.seed(42)
income = np.random.randint(20000, 100000, size=(500,))
spending_score = np.random.randint(1, 100, size=(500,))

# Combine into a dataframe
customer_data = pd.DataFrame({
    'Income': income,
    'Spending_Score': spending_score
})

# Save the synthetic data to a CSV file for further use
customer_data.to_csv('customer_data.csv', index=False)

# Optional: Display the first few rows
print(customer_data.head())
```

```python
# Visualize the data
plt.scatter(customer_data['Income'], customer_data['Spending_Score'])
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Data: Income vs Spending Score')
plt.show()
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(customer_data.isnull().sum())
```

```python
# Summary statistics of the dataset
print("\nSummary statistics:")
print(customer_data.describe())
```

```python
# Feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data)
```

```python
# Implementing K-Means with 3 clusters (3 customer segments)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)
```

```python
# Adding cluster information to the data
customer_data['Cluster'] = kmeans.labels_
```

```python
# Visualize the clusters
plt.scatter(customer_data['Income'], customer_data['Spending_Score'], c=customer_data['Cluster'])
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation with K-means')
plt.show()
```
